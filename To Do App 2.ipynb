```python
#To Do app using Python
# Main menue 
# Add tasks
# View tasks
# Delete tasks
# Terminate app
def isInteger(userInput): # use for desire input from user match your manue list or not. If not not give the error and return false.
    try:
        int(userInput)
        return True
    except ValueError:
        return False
    
    # define Main Manue dictionary values
def mainManue(manue): 
    print("--------------\n  Main Manue     \n--------------\n")
    for manueKey, manueName in manue.items(): # use for loop to print key values and manue names in format list form
        print("{} . {}" .format(manueKey, manueName))
    userCommand = input("\nEnter Command: ")
    return userCommand

# Add new Tasks in dictionary.
def addTask(taskDictionary):
    print("\n--------------\n  Add Tasks     \n--------------\n")
    taskIndex = len(taskDictionary) + 1
    taskName = input("Task # {}" .format(taskIndex))
    taskDictionary .update({taskIndex: taskName})
    print("Task # {} - Successfully Added\n".format(taskIndex))
    
# Viewing all Tasks in Dictionary 
def viewTask(taskDictionary):
    print("\n--------------\n  Viewing Tasks     \n--------------\n")
    if len(taskDictionary) > 0:
        for taskIndex, taskName in taskDictionary.items():
            print("Task #{} - {}" .format(taskIndex,taskName))        
    else:
        print("No Tasks Found\n")
        
# Deleting Tasks according to index number in your Dictionary     
def deleteTask(taskDictionary):
    print("\n--------------\n  Deleting Tasks     \n--------------\n")
    if len(taskDictionary) > 0:
        for taskIndex, taskName in taskDictionary.items():
            print("Task #{} - {}" .format(taskIndex,taskName))
        taskIndex = int(input("Enter Tasks Index eg.1 : "))   
        if taskIndex in taskDictionary.keys():
            taskDictionary.pop(taskIndex)
            print("Task #{} - Succcessfully Deleted" .format(taskIndex))
        else:
            print("Invalid Index, Try Valid Index")
                    
            
    else:
        print("No Tasks Found\n")
        
# Define Main Menue in form of list
manue = {
    1 : "Add Tasks",
    2 : "View Tasks",
    3 : "Delete Tasks", 
    4 : "Terminate App"
}

taskDictionary = {}

while (True):
    userCommand=mainManue(manue)
    
    if isInteger(userCommand) == True: 
        userCommand = int(userCommand)
        if userCommand > 0 and userCommand <=4 :
            # Add Tasks
            if userCommand == 1:
                addTask(taskDictionary)
                
            if userCommand == 2:
                viewTask(taskDictionary)
            
            if userCommand == 3:
                deleteTask(taskDictionary)
                
            if userCommand == 4:
                print("App Terminated")
                break
                
        else:
            print("\nInvalid Command, Type Valid Command")
    else:
            print("\nInvalid Command, Type Valid Command")
    
```
