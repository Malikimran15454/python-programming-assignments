```python
elements = [] #create list
while True: # condition continupusly running while we extrenallay command break the loop
    print("Elements in our list: ")
    for element in elements: # for loop use to print elements one by one while input 1 is running
        print("->" + element) # print all elements which is define from the user
    
    print("\nOptions:\n")
    print("1. Add Elements: ")
    print("2. Delete Elements: ")
    print("3. Delete all Elements: ")
    print("4. Exit: ")
    print(" Note: Please given input in Numbers. eg: 1, 2, 3, 4")
    inp = input("Enter your choice: ") #user define your choice
    if inp == '1':
        elements.append(input("Enter your Element : ")) # user define your elements
    elif inp == '2':
        elements.remove(input("Enter your Element to Delete : ")) # user delete specific value in your list
    
    elif inp == '3':
        elements.clear() # user delete all valuse in your list 
    elif inp == '4':
        break # user break the loop for external command
    else:
        print("Invalid Input")
    print("\nxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx\n")
```
